```python
#!/usr/bin/env python
# coding: utf-8

# <과제4> 2018136085 이세희
# ===

# 1번
# ===

# * 파일 s.txt 내의 다음 내용과 함께 아래 문제의 코드를 제시하시오.

# >pig ham
# 
# >cat dog
# 
# >ham bird
# 
# >dog pig

# 2)
# ---
# 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오.

# In[866]:


file = open("s")
list_1=file.readlines()  #파일을 여는 부분 

list_1[-1]= list_1[-1] + '\n'

list_1.sort()


file_1 = open("s1.txt",'w')
file_1.writelines(list_1)

file.close()
file_1.close()


# <설명> 
# ---
#  먼저 파일을 열고, 라인별로 읽는다. 문제에서는 첫번째 단어에 자체들을 기준으로 라인별 정렬하라고 되어있기 때문이다. 
#  이렇게 값을 받고 Sort를 통해서 값을 정렬 하고, Writelines()를 통해서 그대로 입력해주면 끝이다.
# 
# * 여기서 주의할 점은 파일의 마지막 줄에서는 \n(개행)문자가 없어서 마지막에 \n를 추가했다.  

# 3)
# ---
#  파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오.

# In[138]:


file = open("s")
list_2 = []

list_2=file.readlines()  #파일을 여는 부분 

list_2= sorted(list_2, key=lambda x : x[4])

file_1 = open("s2.txt",'w')
file_1.writelines(list_2)

file.close()
file_1.close()


# <설명>
# ---
# 여기서 내가 해결한 방법은 키를 두번째 단어부터 시작하면 된다. 정렬 키를 두번째 자리 위치인 x[4]를 기반으로 하여 정렬 완료 하였다. 

# 4)
# ---
# 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.
# 

# In[153]:


import re 
file = open("s")
#파일을 여는 부분 


num =  file.readline()

num= num.strip('\n')

list_=num.split(' ')


num =  file.readline()
num= num.strip('\n')
list_.extend(num.split(' '))

num =  file.readline()
num= num.strip('\n')
list_.extend(num.split(' '))

num =  file.readline()
num= num.strip('\n')
list_.extend(num.split(' '))


file_1 = open("s3.txt",'w')

for i in range(2,len(list_),3):
    list_[i]=list_[i] + "\n"
    
list_[0]=list_[0] + " "
list_[1]=list_[1] + " "
list_[3]=list_[3] + " "
list_[4]=list_[4] + " "
list_[6]=list_[6] + " "    




file_1.writelines(list_)

file.close()
file_1.close()


# <설명> 
# ---
# 여기서 내가 구현한 방법은 먼저 s를 순서대로 입력 받고 split을 통해서 각 단어마다 입력 받는다. 
# 
# 그 다음 3개씩 끊어서 각 마지막 단어에 \n 개행 문자를 넣어주고 writelines를 통해서 입력하면,
# 
# 개행문자가 반영되어서 
# 
# 3개씩 텍스트파일에 저장이 된다. 
# 

# 2번
# ===
# >Welcome to Our Service
# 
# >A.Sign Up
# 
# >B.Sign In
# 
# >C.Quit
# 
# 
# 다음과 같은 조건들을 참고하여 회원 가입 및 로그인 프로그램을 작성하시오
# 
# * 1을 선택하면 ID, Password, Name, School의 4가지 정보를 입력받아 파일에 저장하는 프로그램을 작성한다.
# 
# 2-1) 입력된 내용은 access.txt 라는 이름의 텍스트 파일 내에 저장된다.
# 
# 2-2) access.txt 파일의 각 라인에는 가입된 회원 각각의 정보가 "[id]: [password], [name], [school]" 형태로 저장된다.
# 
# 2-3) 즉, 가입 회원이 10명이면 access.txt 파일 내에 라인 수도 정확히 10개이다.
# 
# 2-4) 암호화 방식은 sha 모듈을 활용한다. sha 모듈 활용 방법은 본 문제의 마지막에 제시된 sha 활용 예를 참고한다.
# 
# 즉, access.txt 파일 내에 password 정보는 암호화 되어 저장되어야 한다.
# 
# 2-5) 회원 정보를 입력 받을 때 id를 입력 받은 직후 access.txt를 확인하여 이미 존재하는 id가 입력되었다면 다음 메시지를 출력하고 id 정보를 다시 입력받는다.
# 
# Sorry, the entered ID is already used.
# 
# * 2를 선택하면 ID, Password의 2가지 정보를 입력받는 프로그램을 작성한다.
# 
# 3-1) 입력된 ID 정보가 access.txt에 존재하지 않으면 다음과 같은 메시지를 출력하고 다시 입력받는다.
# 
# Sorry, you are not a registered member.
# 
# 3-2) 입력된 ID가 올바르게 존재하지만 Password 정보가 access.txt 파일에 있는 정보와 불일치하면 다음과 같은 메시지를 출력하고 Password를 다시 입력받는다.
# 
# Sorry, the entered password is not correct.
# 이 때에도 사용자가 입력한 Password 정보와 함께 sha 모듈이 활용되어야 한다.
# 
# 3-3) 입력된 ID와 Password가 모두 올바르면 다음과 같은 메시지를 출력한다.
# Hello [name]!
# 위 [name]에는 access.txt에 기록되어 있는 name 정보를 출력한다.
# 
#  * 3을 선택하면 프로그램이 끝난다.

# In[161]:



import hashlib
import re

file_1 = open("access.txt",'a')   #파일 오픈 
file_2 = open("access.txt",'r')

###########################################################<필요한 변수 목록>
num =  len(file_2.readlines() )    
file_2.seek(0)
trash =''      
list2 = []
trynum=0
id_list=[]
pw_list=[]
##################################################################<메인화면>

print("Welcome to Our Service")
print("A.Sign up")
print("B.Sign in")
print("C.Quit")

choise = input("A,B,C 하나를 고르세요:") 
#####################################################################


#####################################################################<회원가입>

if (choise == 'A'):   
    
    print("ID: ")    #id 입력 받는 곳 
    id = input()
    
    for i in range (num):    #순수 아이디를 추출하기 위한 공간 
        str1 = file_2.readline()
        text_ = str1.replace("[", ":")
        text_ = text_.replace("]", ":")
        text_ = text_.replace(",", ":")  
        list1 = re.split(':',text_)    
  
        while trash in list1:
            list1.remove('') 
            
        id_list.append(list1[0])
    
    while id in id_list :   
        print("Sorry, the entered ID is already used.")
        id = input()
        

            
            
    file_2.close()    #file_2  닫기 
    
    print("Password:  ")    #비번 
    password = input()
    
    print("Name: ")     #이름 
    name = input()
    
    print("School: ")   # 학교 
    school = input()
    
    pw = password    # 암호화 기능 ! 
    pw_encrypted = hashlib.sha256(pw.encode()).hexdigest()
    
    
    
    #여기서 이제  모든 값들을 양식에 맞게 
    txtline = '[{0}]:[{1}],[{2}],[{3}]\n'.format(id,pw_encrypted,name,school)
    
    #text파일에 집어넣는 공간 !
    file_1.write(txtline)  
    
    
    file_1.close()   #file_1 닫기 
    
##############################################################################<로그인 화면>


elif choise == 'B': 
    id_list=[]
    pw_list=[]   
    
    
    print("ID: ")    #id 입력 받는 곳 
    id = input()
    
    for i in range (num):    #순수 아이디를 추출하기 위한 공간 
        str1 = file_2.readline()
        print(str1)
        text_ = str1.replace("[", ":")
        text_ = text_.replace("]", ":")
        text_ = text_.replace(",", ":")
        
        list1 = re.split(':',text_) 
        
        while trash in list1:
            list1.remove('')
 
        id_list.append(list1[0])   #현재 저장된 아이디들을 저장하자 
        
    
        #id가 존재하는지 없는 지 판별 
    while id not in id_list :   
        print("Sorry, you are not a registered member.")
        id = input()    
    
    
    
    
    file_2.seek(0)  #이제 비밀번호 찾을 차례 따라서 read하려면 위치 다시 앞에서 부터 !
    for i in range (num):
        
        str1 = file_2.readline()
        
        text_ = str1.replace("[", ":")
        text_ = text_.replace("]", ":")
        text_ = text_.replace(",", ":")
        
        list1 = re.split(':',text_) 
        
        while trash in list1:
            list1.remove('')
 
        pw_list.append(list1[1])    #현재 저장된 비밀번호들을 저장하자 
    
    
    if(trynum ==0):   #첫시도인가? 
        
        print("Password:  ")    #비번 
        password = input()
            
        trynum = trynum +1 
            
        
        pw = password    # 암호화 기능 ! 
        pw_encrypted = hashlib.sha256(pw.encode()).hexdigest()
            
        while pw_encrypted not in pw_list:    #틀렸다면 반복입력 
            print("Sorry, the entered password is not correct.")
            password = input()    
            pw = password    # 암호화 기능 ! 
            pw_encrypted = hashlib.sha256(pw.encode()).hexdigest()
        
        
        
 
    
    print('Hello [{0}]!'.format(list1[2]))   #로그인 성공시 
    
####################################################################################<종료 화면>    

elif choise == 'C':  
    print("종료합니다")

    
###############################################################################<잘못 입력했을 때>
else:
    print("잘못입력하셨습니다.종료합니다")


# <설명>
# ===
# >이 문제를 만들기 위해서 정말 많은 시간을 투자했다... 
# 
# 이 문제에서 내가 구현한 알고리즘은 다음과 같다. 
# < sign up 경우 >
# 
# * id입력 부분
# 1. txt파일을 한줄씩 가져온다. 
# 2. 순수 id를 추출하기 위해 가져온 한줄을 split을 통해 분해한다. 
# 3. 이제 추출한 id들을 목록에 저장한다. 
# 4. 목록안에 있으면 중복이다. 다시 입력 
# 
# 
# * pw입력 부분 
# 1. 과제에서 제공한 암호화 모듈 사용해서 대입 
# 
# 마지막으로 모든 데이터를 입력 받고 포멧으로 형식 만들어서 write 하기 
# 
# < Sign in 경우 >
# 
# * id입력 부분 (위에서 구현한 것과 매우 비슷하다) 
# 1. txt파일을 한줄씩 가져온다. 
# 2. 순수 id를 추출하기 위해 가져온 한줄을 split을 통해 분해한다. 
# 3. 이제 추출한 id들을 목록에 저장한다. 
# 4. 목록안에 없으면 없는 아이디다 다시 입력 
# 
# * pw입력 부분 (위와 같은 알고리즘 사용) 
# 
# 

# 3번
# ===
# range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
# 

# 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
# 
# >range(stop)
# 
# >range(start, stop[, step])
# 
# 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
# 
# 3) frange 사용 예
# 
# 3-1) frange(0.5)
# > [0.0, 0.1, 0.2, 0.3, 0.4]
# 
# 3-2) frange(1.0, 2.0)
# >[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
# 
# 3-3) frange(2.2, 4.0, 0.5)
# >[2.2, 2.7, 3.2, 3.7]

# In[86]:


def frange(start,stop=-1,step=0.1):
    answer=[]
    if(stop==-1):
        stop = start
        start = 0 
    
    
    while start < stop:
        answer.append(round(start,14))
        start = start + step
        
    return answer
    
print(frange(1.0,2.0))


# <설명> 
# ---
# 여기서 까다로웠던 조건은 defalt 매개 변수는 뒤에서 부터 선언해야하는데, 우리가 매개변수로 주어진것은 stop이 우선이다. 
# 
# 그래서 내가 설정한 조건은 일단 defalt 매개변수의 오류를 막기 위해서 뒤에서 부터 채우고, 음수를 받지 않는다는 조건이 있기 때문에, 
# 
# 일시적으로 stop값에 -1를 대입했다. 
# 
# 그렇다면 def frange(start,stop=-1,step=0.1): 여기서 frange(값1) 이라면 당연히 start = 값1 이 될거지만, 
# 
# if문을 만들어서 stop=-1(stop 값이 안들어왔으면)  이면 start와 stop 값을 변경한다. 

# 4번
# ==
# * 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오
# 
# sum() 사용 예
# > sum()
# 
# > 0
# 
# >sum(1, 2)
# 
# >3
# 
# >sum(1, 2, 3, 4, 5)
# 
# >15
# 
# >sum(1, 5, 7, 2, -10)
# 
# >5

# In[94]:


def sum(*args):
    
    answer_1 = 0
    for arg in args:
        answer_1 = answer_1 + arg
    return answer_1


sum(1,2,3,4,5)


# <설명> 
# --
# 여기서는 가변 인수를 사용하는 법을 익힌다.  *args를 매개변수로 선언하면, 매개변수에 개수를 여러 개 입력받을 수 있다. 
# 
# 반복문을 통해서 args에 있는 값(arg)들을 모두 더해준후 return 하여 마무리한다. 
# 

# 5번
# ==
# * 여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
# 
# 1 다음에 제시되는 함수들을 모두 이용해야 한다.
# >split
# 
# >map
# 
# >join
# 
# 2 myinitial() 함수 사용 예
# >myinitial("as soon as possible")
# 
# >ASAP

# In[105]:


def myinitial(str1):
    list1 = str1.split(' ')
    
    list1 = [i[0].capitalize() for i in list1]
    
    print("".join(map(str,list1)) )
    
    
myinitial("as soon as possible")


# <설명>
# --
# 여기서 3가지 함수를 사용해야 한다. map, split, join
# split으로 단어마다 분해하고 그 단어를 대문자로 만들어야 합니다. 여기서 capitalize()함수를 사용했고 내포를 통해 반복하고 i[0](첫번째 단어만) 저장한다. 
# 
# 마지막으로 A,S,A,P이 리스트에 저장되었는데, join과 map함수를 통해서 이 값들을 짧은 코드로 모두 출력가능하다. 
# 
# "".join(map(str,list1)) 
# 
# 이렇게 구현했다.
# 
# 

# 6번
# --
# 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
# [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.
# 

# In[3]:


import sys
sys.setrecursionlimit(10**7)

def myfact(n):
    if n>1:
        return n*myfact(n-1)
    else :
        return 1

print(myfact(2959))


# <설명>
# ---
# myfact(n-1)을 반복 호출 하면서 n factorial을 구현하였다. 그렇지만, [옵션]에서 설명한 n을 증가하라고 하셨는데, 
# 그래서 1부터 계속 증가해보았다. 그러다가 2959 factorial을 하게 되면, 오류가 발생한다. 
# 
# RecursionError: maximum recursion depth exceeded in comparison
# 오류의 내용은 재귀 호출 함수의 최대호출을 초과했다.
# 
# 여기서 재귀 함수 호출 제한을 풀어주기 위해서는 import sys    sys.setrecursionlimit(10**7)을 통해서 제한을 풀어준다면 더 많은 펙토리얼을 계산할 수 있다. 
# 
# (10000! 을 하니까 터미널이 죽어버린다....) 
# 
# 
# 

# 7번
# =
# (서술형) import string 과 from string import * 의 차이점을 설명하시오.

# >import 모듈
# 
# >from 모듈 import * (전체)
# 
# 입니다. 따라서 String 모듈안에서는 많은 함수들이 구현되어있습니다. 그 함수들을 사용하기 위해서 import string을 했다면 모듈안에 있는 함수들을 string.함수이름 을 통해서 호출 가능합니다.
# 
# from string import 함수이름 으로 선언했다면, 이제부터는 함수이름을 통해 string 모듈안에 있는 함수를 호출할 수 있게됩니다. 
# 여기서 *는 string 모듈안에 있는 모든 함수를 사용하겠다는 의미로 이제부턴, 함수이름을 통해 호출 가능하게 되는 것 입니다. 
# 
# 이 함수 이름들은 이름공간에 저장되어서 사용할 수 있게됩니다.

# [Incremental Project 문제]
# ===

# In[157]:


import requests
import re
import pickle
import operator




file_html = open("www.koreatech.ac.kr.html",'w', encoding="UTF-8-sig")
file_html_1 = open("www.nytimes.com.html",'w', encoding="UTF-8-sig")
file_html_2 = open("www.daum.net.html",'w', encoding="UTF-8-sig")
file_html_3 = open("www.naver.com.html",'w', encoding="UTF-8-sig")
file_html_4 = open("cse.koreatech.ac.kr.html",'w', encoding="UTF-8-sig")





stop_word = open("stop_words_english.txt",encoding='UTF8') #영어 불용어
stop_word_ko = open("stopwords-ko.txt",encoding='UTF8')    #한글 불용어

stop_word.seek(0)            #위치 초기화 
stop_word_ko.seek(0)         #위치 초기화 



source_list = []
source_list_1 = []
source_list_2 = []
source_list_3 = []
source_list_4 = []

req = requests.get('https://www.koreatech.ac.kr') 
req_1 = requests.get('http://www.nytimes.com')       
req_2 = requests.get('http://www.daum.net')       
req_3 = requests.get('http://www.naver.com')       
req_4 = requests.get('http://cse.koreatech.ac.kr')       

source = req.text
source_1 = req_1.text
source_2 = req_2.text
source_3 = req_3.text
source_4 = req_4.text


source = re.sub('<script.*?>.*?</script>', '', source, 0, re.I|re.S)  #Script 사이 내용 삭제 
source =  re.sub(r'(?s)\<.*?\>[^\w\s]*', '', source).replace('\n', '')    # < > 사이 태그 문자 삭제 
source =  re.sub(r'(?s)\>.*?\>[^\w\s]*', '', source).replace('\n', '')     # > > 사이 문자 삭제 
source =  re.sub(r'(?s)\{.*?\}[^\w\s]*', '', source).replace('\n', '')     # {  }사이 문자 제거 
source =  re.sub('[-=+,#/\?:^$.@*\"※◀ⓒ▶~&%ㆍ!』;}\\‘|\(\)\[\]\<\>`\'…》]', '', source)   #특수문자 제거 

source_1 = re.sub('<script.*?>.*?</script>', '', source_1, 0, re.I|re.S)  #Script 사이 내용 삭제 
source_1 =  re.sub(r'(?s)\<.*?\>[^\w\s]*', '', source_1).replace('\n', '')    # < > 사이 태그 문자 삭제 
source_1 =  re.sub(r'(?s)\>.*?\>[^\w\s]*', '', source_1).replace('\n', '')     # > > 사이 문자 삭제 
source_1 =  re.sub(r'(?s)\{.*?\}[^\w\s]*', '', source_1).replace('\n', '')     # {  }사이 문자 제거 
source_1 =  re.sub('[-=+,#/\?:^$.@*\"※◀ⓒ▶~&%ㆍ!』;}\\‘|\(\)\[\]\<\>`\'…》]', '', source_1)   #특수문자 제거 

source_2 = re.sub('<script.*?>.*?</script>', '', source_2, 0, re.I|re.S)  #Script 사이 내용 삭제 
source_2 =  re.sub(r'(?s)\<.*?\>[^\w\s]*', '', source_2).replace('\n', '')    # < > 사이 태그 문자 삭제 
source_2 =  re.sub(r'(?s)\>.*?\>[^\w\s]*', '', source_2).replace('\n', '')     # > > 사이 문자 삭제 
source_2 =  re.sub(r'(?s)\{.*?\}[^\w\s]*', '', source_2).replace('\n', '')     # {  }사이 문자 제거 
source_2 =  re.sub('[-=+,#/\?:^$.@*\"※◀ⓒ▶~&%ㆍ!』;}\\‘|\(\)\[\]\<\>`\'…》]', '', source_2)   #특수문자 제거 

source_3 = re.sub('<script.*?>.*?</script>', '', source_3, 0, re.I|re.S)  #Script 사이 내용 삭제 
source_3 =  re.sub(r'(?s)\<.*?\>[^\w\s]*', '', source_3).replace('\n', '')    # < > 사이 태그 문자 삭제 
source_3 =  re.sub(r'(?s)\>.*?\>[^\w\s]*', '', source_3).replace('\n', '')     # > > 사이 문자 삭제 
source_3 =  re.sub(r'(?s)\{.*?\}[^\w\s]*', '', source_3).replace('\n', '')     # {  }사이 문자 제거 
source_3 =  re.sub('[-=+,#/\?:^$.@*\"※◀ⓒ▶~&%ㆍ!』;}\\‘|\(\)\[\]\<\>`\'…》]', '', source_3)   #특수문자 제거 

source_4 = re.sub('<script.*?>.*?</script>', '', source_4, 0, re.I|re.S)  #Script 사이 내용 삭제 
source_4 =  re.sub(r'(?s)\<.*?\>[^\w\s]*', '', source_4).replace('\n', '')    # < > 사이 태그 문자 삭제 
source_4 =  re.sub(r'(?s)\>.*?\>[^\w\s]*', '', source_4).replace('\n', '')     # > > 사이 문자 삭제 
source_4 =  re.sub(r'(?s)\{.*?\}[^\w\s]*', '', source_4).replace('\n', '')     # {  }사이 문자 제거 
source_4 =  re.sub('[-=+,#/\?:^$.@*\"※◀ⓒ▶~&%ㆍ!』;}\\‘|\(\)\[\]\<\>`\'…》]', '', source_4)   #특수문자 제거 


source_list = source.split()    # 공백으로 문자들을 나누고 
source_list_1 = source_1.split()  
source_list_2 = source_2.split()  
source_list_3 = source_3.split()  
source_list_4 = source_4.split() 


for i in source_list:
    if(len(i)>20):       #띄어쓰기가 안된 엄청 긴 단어들은 표현 x 
        source_list.remove(i)          #주석처리 해제 하면 결과 출력 됩니다! 

for i in source_list_1:
    if(len(i)>20):       #띄어쓰기가 안된 엄청 긴 단어들은 표현 x 
        source_list_1.remove(i)          #주석처리 해제 하면 결과 출력 됩니다!   
        
for i in source_list_2:
    if(len(i)>20):       #띄어쓰기가 안된 엄청 긴 단어들은 표현 x 
        source_list_2.remove(i)          #주석처리 해제 하면 결과 출력 됩니다! 
        
for i in source_list_3:
    if(len(i)>20):       #띄어쓰기가 안된 엄청 긴 단어들은 표현 x 
        source_list_3.remove(i)          #주석처리 해제 하면 결과 출력 됩니다! 
        
for i in source_list_4:
    if(len(i)>20):       #띄어쓰기가 안된 엄청 긴 단어들은 표현 x 
        source_list_4.remove(i)          #주석처리 해제 하면 결과 출력 됩니다!         

        
wordCount = {} 
wordCount_1 = {} 
wordCount_2 = {} 
wordCount_3 = {} 
wordCount_4 = {} 


for word in source_list:      # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴
    wordCount[word] = wordCount.get(word, 0) + 1 
    
for word in source_list_1:      # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴
    wordCount_1[word] = wordCount_1.get(word, 0) + 1     

for word in source_list_2:      # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴
    wordCount_2[word] = wordCount_2.get(word, 0) + 1 
    
for word in source_list_3:      # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴
    wordCount_3[word] = wordCount_3.get(word, 0) + 1 
    
for word in source_list_4:      # Get 명령어를 통해, Dictionary에 Key가 없으면 0리턴
    wordCount_4[word] = wordCount_4.get(word, 0) + 1 
    

    
while(True):   #영어 불용어 처리 

    j = stop_word.readline()
    j = j.strip('\n')
    
    if(j in wordCount):
        del(wordCount[j])
    
    if(j in wordCount_1):
        del(wordCount_1[j])
    
    if(j in wordCount_2):
        del(wordCount_2[j])
        
    if(j in wordCount_3):
        del(wordCount_3[j])
        
    if(j in wordCount_4):
        del(wordCount_4[j])
        
    if j=='':
        break
        
        
        
    
while(True):     #한글 불용어 처리 
    
    k= stop_word_ko.readline()
    k = k.strip('\n')
    
    if(k in wordCount):
        del(wordCount[k])
    
    if(k in wordCount_1):
        del(wordCount_1[k])
    
    if(k in wordCount_2):
        del(wordCount_2[k])
        
    if(k in wordCount_3):
        del(wordCount_3[k])
        
    if(k in wordCount_4):
        del(wordCount_4[k])
    
    if k=='':
        break    
    

    
    


    
    
    
for word in source_list:       #여기서 key정렬 
    keys = sorted(wordCount.keys())    

for word in source_list_1:       #여기서 key정렬 
    keys_1 = sorted(wordCount_1.keys()) 

for word in source_list_2:       #여기서 key정렬 
    keys_2 = sorted(wordCount_2.keys()) 
    
for word in source_list_3:       #여기서 key정렬 
    keys_3 = sorted(wordCount_3.keys()) 
    
for word in source_list_4:       #여기서 key정렬 
    keys_4 = sorted(wordCount_4.keys()) 
    

    
stopword_input=str(input())  #사용자 불용어 처리 
similarity=0
similarity_1=0
similarity_2=0
similarity_3=0
similarity_4=0

while(stopword_input != str(-1)):  
    stopword_input=str(input())
    print(wordCount.keys())
    
    if(stopword_input in wordCount):
        
        similarity=wordCount[stopword_input] /len(source_list) *100
        del(wordCount[stopword_input])
    
    if(stopword_input in wordCount_1):
        
        similarity_1=wordCount_1[stopword_input] /len(source_list_1) *100
        del(wordCount_1[stopword_input])
    
    if(stopword_input in wordCount_2):
        similarity_2=wordCount_2[stopword_input] /len(source_list_2) *100
        del(wordCount_2[stopword_input])
        
    if(stopword_input in wordCount_3):
        similarity_3=wordCount_3[stopword_input] /len(source_list_3) *100
        del(wordCount_3[stopword_input])
        
    if(stopword_input in wordCount_4):
        similarity_4=wordCount_4[stopword_input] /len(source_list_4) *100
        del(wordCount_4[stopword_input])
      
    
    
    
for word in keys:
    inputstring = word + ':' + str(wordCount[word]) 
    file_html.write(inputstring + '\n')                 #파일 저장 ! html 

for word in keys_1:
    inputstring = word + ':' + str(wordCount_1[word]) 
    file_html_1.write(inputstring + '\n') 
    
for word in keys_2:
    inputstring = word + ':' + str(wordCount_2[word]) 
    file_html_2.write(inputstring + '\n')  
    
for word in keys_3:
    inputstring = word + ':' + str(wordCount_3[word]) 
    file_html_3.write(inputstring + '\n')  
    
for word in keys_4:
    inputstring = word + ':' + str(wordCount_4[word]) 
    file_html_4.write(inputstring + '\n')   
    
    
    
    
    
    
    
with open('www.koreatech.ac.kr.pickle','wb') as file_pickle:   #pickle 파일 저장 
    pickle.dump(wordCount,file_pickle) 
    
with open('www.koreatech.ac.kr.pickle', 'rb') as file_pickle_r:  #pickle 파일 로드
    wordCount_p = pickle.load(file_pickle_r)    
      
pickle_result= sorted(wordCount_p.items() , key = lambda item: item[1],reverse=True)   #여기서 빈도수를 계산하여 정렬 
pickle_result= pickle_result[:5]
 ########################################################################################################################
    
    
with open('www.nytimes.com.pickle','wb') as file_pickle_1:   #pickle 파일 저장 
    pickle.dump(wordCount_1,file_pickle_1) 
    
with open('www.nytimes.com.pickle', 'rb') as file_pickle_r_1:  #pickle 파일 로드
    wordCount_p_1 = pickle.load(file_pickle_r_1)    
      
pickle_result_1= sorted(wordCount_p_1.items() , key = lambda item: item[1],reverse=True)   #여기서 빈도수를 계산하여 정렬 
pickle_result_1= pickle_result_1[:5]
#########################################################################################################################

with open('www.daum.net.pickle','wb') as file_pickle_2:   #pickle 파일 저장 
    pickle.dump(wordCount_2,file_pickle_2) 
    
with open('www.daum.net.pickle', 'rb') as file_pickle_r_2:  #pickle 파일 로드
    wordCount_p_2 = pickle.load(file_pickle_r_2)    
      
pickle_result_2= sorted(wordCount_p_2.items() , key = lambda item: item[1],reverse=True)   #여기서 빈도수를 계산하여 정렬 
pickle_result_2= pickle_result_2[:5]

##################################################################################################

with open('www.naver.com.pickle','wb') as file_pickle_3:   #pickle 파일 저장 
    pickle.dump(wordCount_3,file_pickle_3) 
    
with open('www.naver.com.pickle', 'rb') as file_pickle_r_3:  #pickle 파일 로드
    wordCount_p_3 = pickle.load(file_pickle_r_3)    
      
pickle_result_3= sorted(wordCount_p_3.items() , key = lambda item: item[1],reverse=True)   #여기서 빈도수를 계산하여 정렬 
pickle_result_3= pickle_result_3[:5]

####################################################################################################

with open('cse.koreatech.ac.kr.pickle','wb') as file_pickle_4:   #pickle 파일 저장 
    pickle.dump(wordCount_4,file_pickle_4) 
    
with open('cse.koreatech.ac.kr.pickle', 'rb') as file_pickle_r_4:  #pickle 파일 로드
    wordCount_p_4 = pickle.load(file_pickle_r_4)    
      
pickle_result_4= sorted(wordCount_p_4.items() , key = lambda item: item[1],reverse=True)   #여기서 빈도수를 계산하여 정렬 
pickle_result_4= pickle_result_4[:5]




print(pickle_result)
print(pickle_result_1)
print(pickle_result_2)
print(pickle_result_3)
print(pickle_result_4)



file_html.close()
stop_word.close()
file_pickle.close()
file_pickle_r.close()


file_html_1.close()
file_pickle_1.close()
file_pickle_r_1.close()

file_html_2.close()
file_pickle_2.close()
file_pickle_r_2.close()

file_html_3.close()
file_pickle_3.close()
file_pickle_r_3.close()

file_html_4.close()
file_pickle_4.close()
file_pickle_r_4.close()



print("모든 단어의 갯수는 : ",len(source_list), "개 입니다.")   
print("빈도수: ")
print(similarity)
print("빈도수: ")
print(similarity_1)
print("빈도수: ")
print(similarity_2)
print("빈도수: ")
print(similarity_3)
print("빈도수: ")
print(similarity_4)



# <설명>
# ---
# 각 페이지를 html 태그 제거 후 새로운 사전을 만들어서 거기에 빈도수 만큼 key와 값으로 저장하게 되었습니다. 
# 그 것을 파일을 하나씩 저장하고, 그 다음 pickle을 통해서 사전 내용 그대로 파일 저장하고, 
# 다시 피클 파일을 load하여 그 사전들을 sorted를 통해서 빈도수 내림차순으로 저장합니다. 
# 그러면 튜플로 저장되는데, 튜플을 슬라이스를 사용하여 5개만 출력하도록 했습니다. 
# 
# 유사도는 빈도수 / 전체 단어수 * 100 으로 계산했습니다. 
# 
# 
# 
# 

# <소감>
# ===
# [Incremental Project 문제]를 계속해서 부여잡고 해결하려고 하는데, 정말 문제점도 많고 너무 난이도가 어렵다. 많은 시간을 소요했지만, 내가 생각한 알고리즘을 대입할 때 마다 오류가 발생했고, 배우지 않았던 부분까지 구글링하여 검색해야하고 새로운 공부도 필요했기에 더욱 난이도가 있었던 것 같다. 그리고 어떻게 해야할 지 힌트가 없으니 너무 막막하기만 했고, 첫 과제부터 아마 단추가 잘못끼워지지 않았을 까 생각이 든다
# 
# 그래도 이렇게 project문제를 해보니, 정말 대외활동이나, 경진대회에서 나올법한 문제가 나와서 많은 공부가 된 것 같다. 2학년까지 다니면서, 처음으로 프로젝트 문제를 해보는 것 같다. 그래서 더욱 어려운 것 같고, 교재에서 나오지않고 자신이 직접 창의적으로 알고리즘을 구현해야 하는 부분에서 흥미를 느꼈다. 
# 
# 그래도 어려웠고, 이 문제를 해결하기 위해서 계속 계속 노력할 것이다. 
# 
# 
```
