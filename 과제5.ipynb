* 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.


* 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.


* 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.


* 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

```python
class Counter:
    
    count=0
    step=0
      
    def __init__(self, count , step=1):
        self.count = count
        self.step = step 
        
    
    def __str__(self):
        s = '[Count(step:' + str(self.step) + ')]' +" "+ str(self.count)
        return s
    
    def __call__(self):
        self.incr()
        
    def __add__(self,number):
        self.count = self.count + number
        return Counter(self.count,self.step)
        
    def __sub__(self,number):
        self.count = self.count - number
        return Counter(self.count,self.step)
    
    def incr(self):
        self.count =self.count + self.step
        
    def __eq__(self,number):
        return self.count  == number 
    
    def __lt__(self,number):
        return self.count < number 
    def __gt__(self,number):
        return self.count > number 
    
        
#"요구사항1"    
c=Counter(10)
d=Counter(10,2)

print("요구사항2")
print(c)
print(d)

print("요구사항3")
c.incr()
d.incr()
print(c)
print(d)


print("요구사항4")
c()
d()
print(c)
print(d)

print("요구사항5")
c = c + 5
d  = d - 5
print(c)
print(d)

print("요구사항6")
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

* MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
            
        result = result[0:len(result)-2] + "}"  #" ," 없애기 위해서 마지막에 -2를 채워줌 
        
        return result
    
    def __or__(self,other):     # 요구사항1 구현 합집합 구현
        a = []                  #빈 리스트를 만들고 
        for e in self:          #self(s)에 있는 모든 원소 넣기 
            a.append(e)
        for r in other:         #other(t)에 있는 모든 원소 넣기 
            a.append(r)
        MySet.eliminate_duplicate(a) #중복을 제거하면 끝 ! 
        return MySet(a)          #결과를 넣어서 초기값 설정 ! 
    
    def __and__(self,other):       # 요구사항2 구현 교집합 구현
        a = []                     #마찬가지로 빈 리스트 만들고 
        for e in other:
            if e in self:          #other원소 중에 self안에 있다면 둘다 있다는 뜻이므로 그것만 넣어라! 
                a.append(e)
        MySet.eliminate_duplicate(a) #중복제거 
        return MySet(a)              #똑같이 결과 반환 
    
    def __sub__(self,other):           # 요구사항3 구현 교집합 구현        
        for e in other:
            if e in self:               #만약 other원소 중에 self안에 있다면 제거해라! 
                self.remove(e)          #self안에 있는 other원소 e 삭제 ! 
                                        #제거했으니 중복은 없음 
        return MySet(self)              #이번에는 self를 결과로 초기화하면 끝 
        
        

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    

    
```

먼저, __ init __ 메서드부터 확인하자면, __ init __ 은 위에 나온 것 처럼 메서드는 객체가 초기화되고 생성될 때 자동으로 호출 되는 기능을 한다. 여기서는 list를 받고 반복문을 통해서 list를 하나씩 append 넣어줘서 초기화 하고 있다. 

그 다음으로 eliminate_duplicate()를 호출하는데, 
이 함수는 새로운 리스트를 만들어서 if not in 을 통해 중복 값을 제거하는 과정을 거친다. 그런다음 다시 원래 list 변수를 초기화 시키고 다시 담는다. 그러면 결국에는 중복 값은 제거되는 기능을 하는 메서드이다.

마지막으로 __ str() __ 는 사용자가 원하는 형태로 객체 데이터를 담은 문자열 형태로 출력하는데, 여기서 우리가 원하는 형태는 
MySet: {1 ,2 ,3} 이 형태로 출력하려고 하고 있고, 그러기 위해 

1. Myset: { 까지는 그냥 문자로 넣기 
2. 이제 문자열을 모두 출력해야하니까 for문을 통해서 이터레이터 반복을 하고 str() 형변환 해주기
3. 그 다음, 마지막에 반점을 없애주기 위해서 총 길이에서-2까지 슬라이싱으로 출력 (" ," = 2개)
4. 마지막 괄호 닫기 " } "

그리고 @staticmethod def eliminate_duplicate(l) 표시가 있는데, 이 것은 정적 메소드로 클래스 인스턴스를 만들지 않아도 호출이 가능하게 해주는 기능을 한다. 따라서 print( MySet([1, 2, 2, 3]) ) 하여도 정상 작동한다.
